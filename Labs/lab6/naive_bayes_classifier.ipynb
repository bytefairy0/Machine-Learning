```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
iris = load_iris()
```

## Visualization

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
df['target'] = iris.target
df
```

```python
x = df.iloc[:,:-1].values
x
```

```python
y = df.iloc[:,-1].values
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50, shuffle=True)
```

## Naive Bayes Algorithm

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
gnb.score(x_train, y_train)
```

```python
gnb.score(x_test, y_test)
```

```python
y_pred = gnb.predict(x_test)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot(cmap='Blues')
plt.title("Confusion Matris")
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Naive bayes using multinomial NB
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
mnb.score(x_train, y_train)
```

```python
mnb.score(x_test, y_test)
```

```python
# Naive bayes using benoulinb
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
bnb.score(x_train, y_train)
```

```python
bnb.score(x_test, y_test)
```

```python
# Naive bayes algorithm using complement NB
from sklearn.naive_bayes import ComplementNB
cnb = ComplementNB()
cnb.fit(x_train, y_train)
```

```python
cnb.score(x_train, y_train)
```

```python
cnb.score(x_test, y_test)
```

```python
R = {
    "GaussianNB": [gnb.score(x_test,y_test)],
    "MultinomialNB":[mnb.score(x_test,y_test)],
    "BernoulliNB":[bnb.score(x_test,y_test)],
    "ComplementNB":[cnb.score(x_test,y_test)]
}
```
