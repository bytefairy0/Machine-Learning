```python
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from matplotlib.pyplot import rcParams
import matplotlib.pyplot as plt                                                                                                                                    
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
data.head()
```

```python
data['Gender'] = data['Gender'].replace({'Male':0, 'Female':1}, inplace=True)
x = data.iloc[:, 1:4].values
y = data.iloc[:, -1].values
```

```python
x.shape
```

```python
y.shape
```

```python
clf1 = DecisionTreeClassifier()
clf1.fit(x, y)
```

```python
rcParams['figure.figsize'] = 50,30
plot_tree(clf1)
```

```python
x = data.iloc[:, 2:4].values
y = data.iloc[:, -1].values

model = DecisionTreeClassifier(criterion='gini', max_depth=2, random_state=42)
model.fit(x, y)

plt.figure(figsize=(10, 6))
plot_tree(model, feature_names=['Age', 'Estimated Salary'], class_names=['No', 'Yes'], filled=True)
plt.show()
```
